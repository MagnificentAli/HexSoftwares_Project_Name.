```python
# Create a scatter plot between Sleep_Duration and Study_Hours
```

```python
sns.scatterplot(data=student_sleep_patterns, x='Sleep_Duration', y='Study_Hours', hue='Gender')
plt.title('Scatter Plot: Sleep Duration vs. Study Hours by Gender')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Study Hours (hours)')
```

```python
# Create a histogram for 'Sleep_Duration'
```

```python
sns.histplot(student_sleep_patterns['Sleep_Duration'], kde=True, bins=15)

# Add title and labels
plt.title('Histogram: Sleep Duration Distribution')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
```
